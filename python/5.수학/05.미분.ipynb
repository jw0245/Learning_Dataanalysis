```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,y):
    return 2*x**2 + 6*x*y + 7*y**2 - 26*x - 54*y + 107
x = np.linspace(1,16,100)
y = np.linspace(-3,6,90)
x2, y2 = np.meshgrid(x,y) # 3차원 좌표
z = f(x2, y2)

# x로 미분함수
def gx(x, y):
    return 4*x + 6*y - 26
# y로 미분함수
def gy(x, y):
    return 6*x + 14*y - 54

xx = np.linspace(1, 15, 15)
yy = np.linspace(-3,6,10)
xx2, yy2 = np.meshgrid(xx, yy)

GX = gx(xx2, yy2)
GY = gy(xx2, yy2)

plt.figure(figsize=(10,5))
# 등고선
plt.contour(x, y, z, levels=np.logspace(0,3,10))
# 퀴버플랏
plt.quiver(xx2, yy2, GX, GY, color='b')
plt.show()
```

```python
sp = plt.subplot(1,1,1, projection='3d')
sp.plot_surface(x2,y2,z, alpha=0.3, edgecolor='b')
```

```python
def f(x):
    return (x-2)**2 + 2
x = np.linspace(-1,4,100)
plt.plot(x, f(x))
plt.show()
```

```python
def df(x):
    return 2*x - 4
x = np.linspace(-1, 4, 100)
plt.plot(x, f(x),'--')

step = 0.4

# 첫번째
x_ = 0 # x가 0일때
plt.plot(x, df(x_)*x + f(x_), 'k--')
print(df(x_))

# 두번째
x_ = x_ - (step*df(x_))
plt.plot(x, df(x_)*(x-x_) + f(x_), 'r--')
print(df(x_))

# 세번째
x_ = x_ - (step*df(x_))
plt.plot(x, df(x_)*(x-x_) + f(x_), 'g--')
print(df(x_))

plt.ylim(0,10)
plt.show()
```

```python
# optimize
from scipy import optimize
```

```python
x_ = 0
r = optimize.minimize(f, x_)
```

```python
r
```

```python
# 도함수를 직접 넣어주면 계산량이 줄어듦
r = optimize.minimize(f, x_, jac=df)
r
```
