### 파라미터(매개변수) 추가해서 요청
* 기상청 날씨누리 : http://weather.go.kr/weather/main.jsp
* 생활과 산업 > 서비스 > RSS

```python
# 서울경기 날씨
url = 'https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'

# 파라미터 세팅
# 딕셔너리로 선언
# 키 : 파라미터명, 값 : 파라미터값
param = {
    'stnId':'109'
}
# urllib.parse : url에 파라미터 형태로 변환 (a=1&b=2&c=3)
import urllib.parse
parameter = urllib.parse.urlencode(param)
url = url+'?'+parameter

req = urllib.request
res = req.urlopen(url)
data = res.read()
text = data.decode('utf-8')
#text = urllib.request.urlopen(url).read().decode('utf-8')
print(text)
```

```python
# requests 방식
# 서울경기 날씨
url = 'https://www.weather.go.kr/weather/forecast/mid-term-rss3.jsp'

# 파라미터 세팅
# 딕셔너리로 선언
# 키 : 파라미터명, 값 : 파라미터값
param = {
    'stnId':'109'
}
import requests
text = requests.get(url, data=param).text
print(text)
```

### BeautifulSoup
* 파이썬으로 스크래핑을 할 수 있게 해주는 라이브러리
* HTML, XML에서 원하는 정보 추출
* 다운로드 불가
* 아나콘다에 내장

```python
from bs4 import BeautifulSoup
```

```python
# html 예시
html = """
<html>
    <body>
        <h1>스크래핑</h1>
        <div>HTML</div>
        <span>XML</span>
    </body>
</html>
"""
# html 파싱
soup = BeautifulSoup(html) # 객체 생성

# 원하는 부분 추출
text1 = soup.html.body.h1
text2 = soup.html.body.div
text3 = soup.html.body.span
print(text1.string)
print(text2.string)
print(text3.string)
```

* 모든 태그들을 구조적으로 하나하나 선택하는 복잡함
* 원하는 부분만 '선택'해서 추출 (선택자:id, class)

```python
html = """
<html>
    <body>
        <h1 id='title'>데이터 수집</h1>
        <div class='content'>파이썬</div>
        <span>크롤링</span>
    </body>
</html>
"""
soup = BeautifulSoup(html)
# 아이디로 추출
title = soup.find(id='title')
print(title.string)
# 클래스로 추출
contents = soup.find(class_='content') # class 파이썬 예약어
contents.string
```

```python
# 여러개일때 한번에 추출, find_all()
html = """
<html>
    <body>
        <a href='http://www.naver.com' title='네이버'>네이버</a>
        <a href='http://www.daum.net' title='다음'>다음</a>
        <a href='http://www.google.com' title='구글'>구글</a>
    </body>
</html>
"""
soup = BeautifulSoup(html)
for text in soup.find_all('a'):
    print(text.attrs['href'], text.string)
```

#### BeautifulSoup 선택자로 추출하는 메서드
* 하나만 추출 : select_one('선택자')
* 여러개 추출 : select('선택자)

```python
# 윤동주 시인의 작품 중 하늘과 바람과 별과 시 추출 (하나만)
text = soup.select_one('#mw-content-text > div.mw-parser-output > ul:nth-child(6) > li > b > a')
print(text.string)
print(text.get_text())
```

```python
# 윤동주 시인의 모든 작품 추출 (여러개)
text = soup.select('#mw-content-text > div.mw-parser-output > ul:nth-child(6) > li > ul > li > a')
text
```
