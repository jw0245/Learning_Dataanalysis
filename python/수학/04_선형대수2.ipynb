```python
hong = np.array([4,5,2,2])
kim = np.array([4,0,2,0])
lee = np.array([2,2,0,1])
```

```python
# hong, kim의 유클리디안거리
np.sqrt(np.linalg.norm(hong)**2 + np.linalg.norm(kim)**2 + -2*hong.T@kim)
```

```python
# kim. lee 유클리디언거리
np.sqrt(np.linalg.norm(kim)**2 + np.linalg.norm(lee)**2 + -2*kim.T@lee)
```

```python
# hong. lee 유클리디언거리
np.sqrt(np.linalg.norm(hong)**2 + np.linalg.norm(lee)**2 + -2*hong.T@lee)
```

```python
# hong, kim의 코사인 거리
1-((hong.T@kim)/(np.linalg.norm(hong)*np.linalg.norm(kim)))
```

```python
# kim, lee의 코사인 거리
1-((kim.T@lee)/(np.linalg.norm(kim)*np.linalg.norm(lee)))
```

```python
# hong, lee의 코사인 거리
1-((hong.T@lee)/(np.linalg.norm(hong)*np.linalg.norm(lee)))
```

```python
A = np.array([
    [1,5,6],
    [2,6,8],
    [3,11,14],
    [1,4,5]
])
B = np.array([
    [1,5,6],
    [2,6,8],
    [3,11,14],
    [1,4,8]
])
np.linalg.matrix_rank(A), np.linalg.matrix_rank(B)
```

```python
# 풀랭크가 아니다 -> 회귀분석이 잘안된다, 역행렬이 안만들어짐
np.linalg.inv(A)
```

```python
A = np.array([
    [1,5,6],
    [2,6,8],
    [3,11,5]
])
# 역행렬이 존재하는지 여부
np.linalg.det(A) # 0이면 역행렬이 존재하지 않음
```

```python
np.linalg.inv(A)
```

```python
np.linalg.matrix_rank(A) # 풀랭크
```
