```python
# heaviside step
def hsf(x):
    if x >= 0:
        return 1
    else:
        return 0
hsf(10), hsf(-5)
```

```python
# x>0 조건하에
# 함수f와 f의 역함수 f1
def f(x):
    return x**2
def f1(x):
    return np.sqrt(x)
```

```python
f(2), f1(4)
```

```python
f(np.array([1,2,3])), f1(np.array([1, 4, 9]))
```

```python
# 지수함수
x = np.linspace(-2,2,100)
y = np.exp(x)
plt.plot(x, y)
plt.axhline(1, c='red', ls='--')
plt.axvline(0, c='red', ls='--')
plt.show()
```

```python
# 로그함수
x = np.linspace(0.01, 10, 100)
y = np.log(x)
plt.plot(x, y)
plt.show()
```

```python
# 어떤 함수에 로그를 적용해도 함수의 최고점, 최저점의 위치는 변하지 않음
def f(x):
    return x**3 - 12*x + 20*np.sin(x) + 7

x = np.linspace(-4,4,300)
y = f(x)

plt.subplot(211)
plt.plot(x, y)
plt.subplot(212)
plt.plot(x,np.log(y))
plt.show()
```

```python
# 시그모이드함수
def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

```python
sigmoid(0.8)
```

```python
x = np.linspace(-10,10,100)
plt.plot(x, sigmoid(x))
plt.show()
```

```python
# relu 함수
x = np.linspace(-10,10,100)
plt.plot(x, np.maximum(x,0))
plt.show()
```

```python
# softplus 함수
def softplus(x):
    return np.log(1+np.exp(x))
x = np.linspace(-10,10,100)
plt.plot(x, softplus(x))
plt.show()
```

```python
# 2차원 다변수 함수
def f(x,y):
    return 2*x**2 + 6*x*y + 7*y**2 - 26*x - 54*y + 107
```

```python
x = np.linspace(-3,7,100)
y = np.linspace(-3,7,100)
x2, y2 = np.meshgrid(x,y)
z = f(x2,y2)

fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
ax.plot_surface(x2,y2,z)
ax.view_init(40,-110)
plt.show()
```

```python
plt.contour(x2,y2,z, levels=np.logspace(0,3,10))
plt.show()
```

```python
# 소프트맥스 함수
def softmax(x,w) :
    # x:1차원 벡터, w:가중치 벡터
    return np.exp(w*x) / np.exp(w*x).sum()
```

```python
x = [2,1,0.5]
y = softmax(x, np.ones(3))
y
```

```python
np.sum(y)
```

```python
x = np.linspace(-10,10,100)
plt.plot(x, sigmoid(x))
plt.plot(x, sigmoid(x+2)-1)
plt.show()
```

```python
x = np.linspace(-10,10,100)
# 시그모이드 함수를 x축 1/2배 y축 2배
plt.plot(x, sigmoid(x))
plt.plot(x, 2*sigmoid(2*x))
plt.show()
```

```python
# tanh : 하이퍼블릭 탄젠트 함수
plt.plot(x, sigmoid(x))
plt.plot(x, 2*sigmoid(2*x)-1)
plt.show()
```
