### pandas 내장 그래프
* plot() 메서드를 사용
* 매개변수로 kind라는 옵션을 통해 그래프 종류 지정
* 그래프 종류
    * line : 선그래프,라인차트 (기본값)
    * bar : 수직 막대그래프
    * barh : 수평 막대그래프
    * hist : 히스토그램
    * box : 박스플롯
    * kde : 커널밀도그래프(히스토그램을 부드럽게 그려준 그래프)
    * area : 면적 그래프
    * pie : 파이 그래프
    * scatter : 산점도 그래프
    * hexbin : 고밀도 산점도 그래프

```python
# 시도별 전출입데이터 불러오기
df = pd.read_excel('../data/시도별 전출입 인구수.xlsx')
df.head()
```

```python
# 결측치 처리
# 앞에 데이터로 채우기
df = df.fillna(method='ffill')
df
```

```python
# 서울에서 타지역으로 이동한 데이터만 추출
# 전출지가 서울, 전입지가 서울이아니고, 전입지가 전국이 아니고
df_seoul = df[(df['전출지별']=='서울특별시')&(df['전입지별']!='서울특별시')&(df['전입지별']!='전국')]
```

```python
df_seoul
```

```python
# 전출지별 삭제
df_seoul = df_seoul.drop('전출지별', axis=1)
df_seoul
```

```python
# 전입지별 -> 전입지 컬럼명 변경
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)
df_seoul
```

```python
# 전입지 열을 인덱스(행이름)로 지정
df_seoul.set_index('전입지', inplace=True)
```

```python
df_seoul
```

```python
# 경기도로 이동한 데이터
df_seoul.loc['경기도'].plot()
```

```python
s_g = df_seoul.loc['경기도']
#plt.plot(s_g)
plt.plot(s_g.index, s_g.values)
```

```python
s_g
```

```python
# figure 사이즈
plt.figure(figsize=(10,6))

# 차트그리기
plt.plot(s_g)

# x축 이름 겹치는 문제
plt.xticks(rotation='90')

# 제목
plt.title('서울->경기도 이동 인구수')
# x축label
plt.xlabel('연도')
# y축label
plt.ylabel('인구수')
# 범례
plt.legend(labels=['서울->경기'])

plt.show()
```

```python
# 스타일(테마) 지정
plt.style.use('ggplot')

# figure 사이즈
plt.figure(figsize=(10,6))

# 차트그리기
plt.plot(s_g, marker='o', color='blue')

# x축 이름 겹치는 문제
plt.xticks(rotation='90')

# 제목
plt.title('서울->경기도 이동 인구수')
# x축label
plt.xlabel('연도')
# y축label
plt.ylabel('인구수')
# 범례
plt.legend(labels=['서울->경기'])

plt.show()
```

```python
# 스타일(테마)의 종류
plt.style.available
```

```python
# 그래프 여러개 그리기

# figure 설정
fig = plt.figure(figsize=(10,10))
# 그래프 갯수만큼 axe 지정
ax1 = fig.add_subplot(2,1,1) # 행,열,순서
ax2 = fig.add_subplot(2,1,2)

# 첫번째 차트그리기
ax1.plot(s_g, marker='o', color='red')

# 두번째 차트
ax2.plot(s_g, marker='o', color='blue')

# x축 이름 겹치는 문제
ax1.set_xticklabels(labels=s_g.index, rotation='90')
ax2.set_xticklabels(labels=s_g.index, rotation='90')

# 제목
ax1.set_title('서울->경기도 이동 인구수')
ax2.set_title('서울->경기도 이동 인구수')
# x축label
plt.xlabel('연도')
# y축label
plt.ylabel('인구수')
# 범례
plt.legend(labels=['서울->경기'])

plt.show()
```

```python
# 하나의 axe객체에 여러개의 그래프 추가
# 서울에서 충청남도, 경상북도, 강원도로 이동한 인구수 추출
df3 = df_seoul.loc[['충청남도','경상북도','강원도'],:]
df3
```

```python
# 영역지정
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1)

# ax객체에 3개의 plot
ax.plot(df3.columns, df3.loc['충청남도',:], color='blue', label='서울->충남')
ax.plot(df3.columns, df3.loc['경상북도',:], color='green', label='서울->경북')
ax.plot(df3.columns, df3.loc['강원도',:], color='red', label='서울->강원')

ax.legend(loc='best')
ax.set_title('서울 ->충남/경북/강원 인구 이동')
ax.set_xlabel('연도')
ax.set_ylabel('인구수')

plt.show()
```
