```python
x = np.arange(0,10,0.1)
y = 1+np.sin(x)
plt.plot(x,y)
```

```python
# darkgrid, whitegrid, white, dark...
sns.set_style('whitegrid')
sns.lineplot(x,y)
```

```python
y2 = 1+np.cos(x)
sns.lineplot(x,y,color='r',alpha=0.7, marker='o')
sns.lineplot(x,y2,color='g',alpha=0.7, marker='^')
plt.show()
```

```python
# piechart는 seaborn에 없음
# startangle : 시작각도
# explode : 한조각 튀어나오는 비율
# shadow : 그림자
# autopct : 퍼센트 자동으로 표기 %포맷%%

# 가상의 데이터(스마트폰 점유율)
company = ['삼성','애플','LG','화웨이','샤오미']
values = [40,30,20,7,3]
color = ['red','green','yellow','purple','gray'] # 색상
plt.figure(figsize=(10,5))
plt.pie(values, labels=company, startangle=90, explode=[0.2,0,0,0,0], shadow=True, 
        autopct=lambda x:'{:.1f}%'.format(x), colors=color)

plt.legend(loc='best')
plt.show()
```
