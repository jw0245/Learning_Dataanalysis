### 원핫인코딩(one-hot-encoding)
* 혈액형을 그대로 머신러닝 알고리즘으로 예측하게 되면, 학습할때 컴퓨터가 값들간에 관계를 스스로 형성하게 됨
* 예) B형은1, AB형은2 그래서 B형+AB형 = O형
* 별도의 컬럼을 만들어주고 각 값들을 True,False로 넣어줌 (반드시 하나만 True이고 나머지는 False)
* 혈액형들의 관계는 독립적이다

```python
df['혈액형'].value_counts()
```

```python
blood_map = {
    'A':0,
    'B':1,
    'AB':2,
    'O':3
}
df['혈액형_코드'] = df['혈액형'].map(blood_map)
```

```python
df['혈액형_코드'].value_counts()
```

```python
df
```

```python
df2 = pd.get_dummies(df['혈액형'])
df2.columns = ['혈액형_A','혈액형_AB', '혈액형_B', '혈액형_O']
df2
```

```python
pd.concat([df,df2],axis=1)
```

### 파이썬 성능향상

```python
df = pd.DataFrame({'a':[10,20,30],'b':[20,30,40]})
df
```

```python
def avg(r):
    return (r[0]+r[1])/len(r)
```

```python
%%timeit
df.apply(avg, axis=1)
```

```python
# 넘파이로 벡터화한 함수
import numpy as np
@np.vectorize
def avg2(x,y):
    return (x+y)/2
```

```python
%%timeit
avg2(df['a'],df['b'])
```

```python
# numba 라이브러리
import numba
@numba.vectorize
def avg3(x,y):
    return (x+y)/2
```

```python
%%timeit
avg3(df['a'].values,df['b'].values)
```
